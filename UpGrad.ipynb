```python
#converting String to Numeric Values
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df[["job"]] = lb.fit_transform(np.array(df[["job"]]))
df[["marital"]] = lb.fit_transform(np.array(df[["marital"]]))
df[["education"]] = lb.fit_transform(np.array(df[["education"]]))
df[["default"]] = lb.fit_transform(np.array(df[["default"]]))
df[["housing"]] = lb.fit_transform(np.array(df[["housing"]]))
df[["loan"]] = lb.fit_transform(np.array(df[["loan"]]))
df[["contact"]] = lb.fit_transform(np.array(df[["contact"]]))
df[["day"]] = lb.fit_transform(np.array(df[["day"]]))
df[["month"]] = lb.fit_transform(np.array(df[["month"]]))
df[["poutcome"]] = lb.fit_transform(np.array(df[["poutcome"]]))
df[["y"]] = lb.fit_transform(np.array(df[["y"]]))
```

```python
df.head()
```

```python
#dropping day Columns
df = df.drop(["day"], axis=1)
```

```python
df.head()
```

```python
from sklearn import model_selection
from pandas.plotting import scatter_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
models = [("LR",LogisticRegression()),("KNN",KNeighborsClassifier()),
          ("DT",DecisionTreeClassifier()),
          ("RFC",RandomForestClassifier(n_estimators=400,min_samples_split=4)),("xg",XGBClassifier())]
```

```python
x = df.iloc[:,:15]
y = df.iloc[:,15:16]
```

```python
# Standardising the data.
from sklearn.preprocessing import scale

for i in range(0,15):
    x.iloc[:,i] = scale(x.iloc[:,i])
```

```python
# splitting the Data into Training and Test Data
seed = 7
validation_size = 0.2
x_train,x_test,y_train,y_test = model_selection.train_test_split(x,y,test_size = validation_size, random_state = seed)
```

```python
x.head()
```

```python
for name,model in models:
    k_fold = model_selection.KFold(n_splits=10)
    cv_results = model_selection.cross_val_score(model,x_train,y_train,cv=k_fold,scoring="accuracy")
    msg = "%s:%f, %f" % (name,cv_results.mean(),cv_results.std())
    print(msg)
```

```python
for name,model in models:
    print(name)
    model.fit(x_train,y_train)                   #model Fit
    predictions = model.predict(x_test)
    print(accuracy_score(y_test,predictions))    #printing Accuracy Score 
    print(classification_report(y_test,predictions))
```
